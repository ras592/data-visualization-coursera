# Library Imports and Configurations

```python
import altair as alt
import pandas as pd
```

```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
alt.renderers.enable('notebook')
```

# Import Data and Wrangle

```python
df = pd.read_csv('https://data.giss.nasa.gov/gistemp/tabledata_v4/ZonAnn.Ts+dSST.csv')
df.index = df['Year']
_df = df[['Glob', 'NHem', 'SHem']].stack().reset_index().rename({'level_1': 'Temp Type', 0: 'Annual Temp Deviations'}, axis=1)
```

# Data Visualization

```python
chart = alt.Chart(_df).mark_line().encode(
    x=alt.X(
        'Year:N'
    ),
    y='Annual Temp Deviations:Q',
    color=alt.Color('Temp Type:N', legend=alt.Legend(orient="left")),
    tooltip=['Year', 'Annual Temp Deviations', 'Temp Type']
).properties(
    title='Zonal Annual Means (1880-present)'
)

chart = chart.configure_title(fontSize=20, offset=5, orient='top', anchor='start')

# display the chart
chart
```

# Question Response

I chose to remove the latitude mean annual temperature deviation values and focus on the global, Northern hemisphere, Southern hemisphere mean.
